```python
from IPython.display import display
import pandas as pd
import _sqlite3

aux = pd.DataFrame()

def get_compact_df(path, drop_cols):
    """ Função para importar um DataFrame de um arquivo csv, permite tratamento das DFs
    :param path: Caminho do qual se deve importar
    :param drop_cols: Colunas a remover
    :return: se a importação deu certo, retorna a DF importada e tratada, caso contrário retorna uma DF vazia
    """
    
    try:
        df = pd.read_csv(path, index_col=0)
        df.drop(drop_cols, axis=1, inplace=True)
        df.dropna(how='any', axis=0, inplace=True)
        df.dropna(how='all', axis=1, inplace=True)
    except (FileNotFoundError, AttributeError): return pd.DataFrame()
    else: return df
```

## Neste projeto, trabalharemos com dois arquivos csvs separados, contendo informações sobre o preço da gasolina no Brasil. 

### ***1. Carregue os conjuntos de dados "gasolina_2000+.csv" e "gasolina_2010+.csv" em dois DataFrames diferentes e combine-os em um único DataFrame.***

```python
useless_cols = ['MARGEM MÉDIA REVENDA', 'PREÇO MÉDIO DISTRIBUIÇÃO', 'DESVIO PADRÃO DISTRIBUIÇÃO',  'PREÇO MÍNIMO DISTRIBUIÇÃO',  'PREÇO MÁXIMO DISTRIBUIÇÃO',  'COEF DE VARIAÇÃO DISTRIBUIÇÃO']

""" ## Importação de DataFrames """
df1 = get_compact_df("..\\..\\Docs\\projetos\\analise-precos-gasolina\\gasolina_2000+.csv", useless_cols)
df2 = get_compact_df("..\\..\\Docs\\projetos\\analise-precos-gasolina\\gasolina_2010+.csv", useless_cols)

""" ## Verifica situação da Importação """
if (df1.empty) | (df2.empty): print('Tentativa de importar DF vazia ou Falha ao importar. Sistema Encerrado!')
else:
    gas_df = pd.concat([df1, df2], ignore_index=True)
    display(gas_df.head())
```

### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
""" ## Filtrando um DataFrame apenas com dados que serão úteis """
reg_gas_df['MES'] = reg_gas_df['DATA FINAL'].apply(lambda x: x.month)
df_rio = reg_gas_df[reg_gas_df['ESTADO'] == 'RIO DE JANEIRO']

""" ## Pegar apenas a última entrada de cada mês """
df_month_rio = df_rio.groupby('ANO/MES')[['PREÇO MÉDIO REVENDA', 'MES']].last()

""" ## Pegar apenas o último mês do ano """
df_dec_rio = df_month_rio[df_month_rio['MES'] == 12]

""" ## Aplicar shift para poder obter um percentual de mudança do ano atual para o valor anterior (ano passado) """
aux['RAZÃO DO PREÇO A CADA ANO'] = (df_dec_rio['PREÇO MÉDIO REVENDA'] / df_dec_rio['PREÇO MÉDIO REVENDA'].shift(1) - 1) * 100

""" ## Aplicar rolling e sum para obter somatório dos percentuais dos últimos 3 anos """
aux['PERCENTUAL ULTIMOS 3 ANOS'] = aux['RAZÃO DO PREÇO A CADA ANO'].rolling(3).sum()

display(aux)
```

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
""" ## Obter DataFrames com valores úteis """
min_df = reg_gas_df.groupby('ANO/MES').min() ['PREÇO MÉDIO REVENDA']
max_df = reg_gas_df.groupby('ANO/MES').max() ['PREÇO MÉDIO REVENDA']


""" ## Criação de um novo DataFrame com seus respectivos campos e valores """
diff_df = pd.DataFrame()
diff_df['MAX PRICE'] = max_df
diff_df['MIN PRICE'] = min_df
diff_df['ABS DIFF'] = max_df - min_df
diff_df['REL DIFF'] = (diff_df['ABS DIFF'] / reg_gas_df.groupby('ANO/MES')['PREÇO MÉDIO REVENDA'].mean()) * 100

""" ## Captar os indexes de cada elemento para cada valor máximo/mínimo """
idx_max_df = reg_gas_df.groupby('ANO/MES') ['PREÇO MÉDIO REVENDA'].idxmax()
idx_min_df = reg_gas_df.groupby('ANO/MES') ['PREÇO MÉDIO REVENDA'].idxmin()

""" ## Acessar Estados de ocorrência dos valores """
diff_df['STATES MAX'] = reg_gas_df.loc[idx_max_df, :] ['ESTADO'].values
diff_df['STATES MIN'] = reg_gas_df.loc[idx_min_df, :] ['ESTADO'].values

aux = diff_df.copy()
display(aux)
```
